<h3>3.(б). Модель для розпізнавання іменованих сутностей (NER): </h3>

Код нижче демонструє використання української моделі для розпізнавання іменованих сутностей (NER) із бібліотеки Hugging Face.

```python
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification

# Завантаження токенайзера та моделі
tokenizer = AutoTokenizer.from_pretrained('ukr-models/uk-ner')
model = AutoModelForTokenClassification.from_pretrained('ukr-models/uk-ner')

# Створення пайплайну для NER
ner = pipeline('ner', model=model, tokenizer=tokenizer)

# Текст для аналізу
text = "Могила Тараса Шевченка знаходиться в місті Канів."

# Розпізнавання іменованих сутностей
entities = ner(text)
print(entities)
```

3.(а) Класифікація текстів

```python
# === 1. Підготовка ===============================================
import pandas as pd
from transformers import pipeline

# Завантажимо невеликий піднабір твоїх текстів
url = "https://raw.githubusercontent.com/ilchukjulia059-cyber/NeuroLabs/refs/heads/main/Lab2/df_file.csv"
df = pd.read_csv(url)

# Візьмемо перші 100 текстів для демонстрації
df_demo = df.head(100)
print(f"Використовуємо {len(df_demo)} текстів для демонстрації:")
print(df_demo.head(5))

# === 2. Ініціалізація pipeline ===================================
# Використаємо готову модель для класифікації тональності (Sentiment Analysis)
classifier = pipeline("sentiment-analysis")  # автоматично завантажить distilbert-base-uncased-finetuned-sst-2-english

# === 3. Демонстрація передбачень =================================
results = []
for text in df_demo["Text"]:
    pred = classifier(text[:512])[0]  # беремо лише перші 512 токенів, щоб уникнути помилок
    results.append({"text": text, "label": pred["label"], "score": pred["score"]})

# Перетворимо у DataFrame для зручного перегляду
df_results = pd.DataFrame(results)
print("\nПерші 10 передбачень:")
print(df_results.head(10))

# === 4. Простий підрахунок частоти передбачених класів ==========
print("\nЧастота передбачених класів:")
print(df_results["label"].value_counts())
```

*Примітка щодо вибору моделі для демонстрації:*

Через невідповідність класів у моєму датасеті з лабораторної роботи №2 (тексти мали власні категорії, які не підходили до готових Hugging Face моделей), навчання моделей для класифікації давало нульову точність.
Тому для демонстрації роботи бібліотеки Hugging Face я використала готовий pipeline sentiment-analysis (модель distilbert-base-uncased-finetuned-sst-2-english), який класифікує тексти на два класи: POSITIVE / NEGATIVE.
Для економії часу та уникнення зависання було обрано підмножину з 100 текстів. Цей приклад показує працездатність підходу на практиці.
